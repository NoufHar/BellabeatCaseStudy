## <font color='#E2A76F'>5. ANALYZE Phases</font> <a class="anchor" id="analyze"> </a>

Uploading the Excel files to SQL

While cleaning, I noticed that the dailyactivity, dailycalories, dailyintensities, and dailysteps tables had similar columns, so I used an INNER JOIN statement to make sure the data matched based on user IDs and activity dates, And I ran the following query.


```python
SELECT activity.calories,
calories.calories
FROM [bbt].[dailyactivity_merged] activity
INNER JOIN [bbt].[dailycalories_merged] calories
ON activity.id = calories.id 
AND activity.ActivityDate = calories.ActivityDate
```

```python
SELECT activity.SedentaryMinutes,
intensities.SedentaryMinutes
FROM [bbt].[dailyactivity_merged] activity
INNER JOIN [bbt].[dailyIntensities_merged] intensities
ON activity.id = intensities.id 
AND activity.ActivityDate = intensities.ActivityDate
```

```python
SELECT activity.TotalSteps,
steps.StepTotal
FROM [bbt].[dailyctivity_merged] activity
INNER JOIN [bbt].[dailysteps_merged] steps
ON activity.id = steps.id 
AND activity.ActivityDate = steps.ActivityDate
```

The result showed that the daily calories, daily intensity, and daily steps tables contain the same data as the daily activity table, so we will stick with the daily activity table only and drop the others.

First, I wanted to know how many times each ID logged in/worn their Fitbit watches, so I queried the data using

```python
SELECT id,
COUNT(id) AS total_id
FROM [bbt].[dailyactivity_merged]
GROUP BY id
```

The result showed that dailyactivity for each user-id ranged from 4 to 31 times so I went ahead to classify their activities into three classes which are

* Active users.
* Moderate users.
* Light users.

```python
SELECT id,
COUNT(id) AS total_uses,
CASE
WHEN COUNT(Id) BETWEEN 21 AND 31 THEN 'active_user'
WHEN COUNT(Id) BETWEEN 11 and 20 THEN 'moderate_user'
WHEN COUNT(Id) BETWEEN 0 and 10 THEN 'light_user'
END AS user_classification
FROM [bbt].[dailyactivity_merged]
GROUP BY id
```

This showed that 21.2% of users are light users the remaining 78.8% are moderate users but none fell within the active category.

Next, I took a look at the minimum, maximum, and average of the total steps, calories, total distance, and other activity levels represented in the daily activity table using the following query

```python
SELECT Id,
ROUND(MIN(TotalSteps),2) AS min_total_steps,
ROUND(MAX(TotalSteps),2) AS max_total_steps, 
ROUND(AVG(TotalSteps),2) AS avg_total_steps,
ROUND(MIN(TotalDistance),2) AS min_total_dist, 
ROUND(MAX(TotalDistance),2) AS max_total_dist, 
ROUND(AVG(TotalDistance),2) AS avg_total_dist,
ROUND(MIN(Calories),2) AS min_calories,
ROUND(MAX(Calories),2) AS max_calories,
ROUND(AVG(Calories),2) AS avg_calories,
ROUND(MIN(VeryActiveMinutes),2) AS min_very_active_minutes,
ROUND(MAX(VeryActiveMinutes),2) AS max_very_active_minutes,
ROUND(AVG(VeryActiveMinutes),2) AS avg_very_active_minutes,
ROUND(MIN(FairlyActiveMinutes),2) AS min_fairly_active_minutes,
ROUND(MAX(FairlyActiveMinutes),2) AS max_fairly_active_minutes,
ROUND(AVG(FairlyActiveMinutes),2) AS avg_fairly_active_minutes,
ROUND(MIN(LightlyActiveMinutes),2) AS min_lightly_active_minutes,
ROUND(MAX(LightlyActiveMinutes),2) AS max_lightly_active_minutes,
ROUND(AVG(LightlyActiveMinutes),2) AS avg_lightly_active_minutes,
ROUND(MIN(SedentaryMinutes),2) AS min_sedentary_minutes,
ROUND(MAX(SedentaryMinutes),2) AS max_sedentary_minutes,
ROUND(AVG(SedentaryMinutes),2) AS avg_sedentary_minutes
FROM [bbt].[dailyactivity_merged]
GROUP BY id
ORDER BY id
```

Then I wanted to take a look at the highest average number of minutes spent actively

```python
SELECT id,
ROUND(Avg(veryactiveMinutes),2) AS avg_very_active_minutes,
ROUND(Avg(fairlyactiveMinutes),2) AS avg_fairly_active_minutes,
ROUND(Avg(lightlyactiveMinutes),2) AS avg_lightly_active_minutes,
ROUND(Avg(sedentaryminutes),2) AS avg_sedentary_minutes
FROM [bbt].[dailyactivity_merged]
GROUP BY Id
ORDER BY 2,3,4,5 DESC
```

By going back to pedometer indices for public health, I will be using it to classify the total steps for each user id. 

Where it  proposed the following:

* Sedentary lifestyle index < 5000 steps a day
* low-active 5000 ≤ 7499
* somewhat active 7500 ≤ 9999
* Active > 10000
* Highly active >12500

```python
SELECT Id, ROUND(AVG(TotalSteps),2) AS avg_totalsteps,
  CASE
  WHEN ROUND(AVG(TotalSteps),2) < 5000 THEN 'sedentary_lifestyle'
  WHEN ROUND(AVG(TotalSteps),2) BETWEEN 5000 and 7499 THEN 'low_active'
  WHEN ROUND(AVG(TotalSteps),2) BETWEEN 7500 and 9999 THEN 'somewhat_active'
  WHEN ROUND(AVG(TotalSteps),2) BETWEEN 10000 and 12499 THEN 'active'
  WHEN ROUND(AVG(TotalSteps),2) > 12499 THEN 'highly_active'
  END AS total_steps_indices
FROM [bbt].[dailyActivity_merged]
Group by Id
ORDER BY avg_totalsteps
```

Returning to the World Health Organization's recommendations that adults aged 18 to 64 years should engage in at least 150 to 300 minutes of moderate-intensity aerobic physical activity.

to verify if users were meeting up with the WHO recommendation of 150–300 i queried :

```python
 SELECT Id, 
avg(veryactiveminutes) + avg(fairlyactiveMinutes) + avg(lightlyactiveMinutes) AS total_avg_active_minutes,
CASE 
WHEN avg(veryactiveminutes) + avg(fairlyactiveminutes) + avg(lightlyactiveminutes) >= 300 THEN 'exceeds_WHO_recommendation'
WHEN avg(veryactiveminutes) + avg(fairlyactiveminutes) + avg(lightlyactiveminutes) BETWEEN 150 AND 300 THEN 'meets_WHO_recommendation'
WHEN avg(veryactiveminutes) + avg(fairlyactiveminutes) + avg(lightlyactiveminutes) <150 THEN 'does_not_meets_WHO_recommendation'
END 'WHO recommendation'
FROM [bbt].[dailyactivity_merged]
GROUP BY Id
```

The above query showed that 7 users fell under the sedentary lifestyle category
11 users were classified as low active
7 users as somewhat active
6 as active and
2 as highly active

The result showed that the sedentary minutes column was the highest on average.

I repeated the same query but substituted the id column for day_of_the_week because I wanted to take a look at the day with the highest average minutes and the result indicated Monday with sedentary minutes maintaining the highest position on average.

```python
SELECT day_of_the_week,
ROUND(Avg(veryactiveMinutes),2) AS avg_very_active_minutes,
ROUND(Avg(fairlyactiveMinutes),2) AS avg_fairly_active_minutes,
ROUND(Avg(lightlyactiveMinutes),2) AS avg_lightly_active_minutes,
ROUND(Avg(sedentaryminutes),2) AS avg_sedentary_minutes
FROM [bbt].[dailyactivity_merged]
GROUP BY day_of_the_week
ORDER BY 2,3,4,5 DESC
```

The result of this query showed that only 5 out of the 33 users did not meet the recommendation, 2 users exceeded the recommendation and 26 users met the recommendation.

I decided to check out what happens when I remove the lightly active minutes

```python
SELECT Id, 
avg(veryactiveminutes) + avg(fairlyactiveMinutes) AS total_avg_active_minutes,
CASE 
WHEN avg(veryactiveminutes) + avg(fairlyactiveminutes) >= 150 THEN 'meets_WHO_recommendation'
WHEN avg(veryactiveminutes) + avg(fairlyactiveminutes) <150 THEN 'does_not_meets_WHO_recommendation'
END 'WHO recommendation'
FROM [bbt].[dailyactivity_merged]
GROUP BY Id
```

found out that non of the users were able to meet up with the WHO’s minimum recommendation of 150 minutes.

Then i wanted to look at the average of all the major activities recorded in the daily activity table by day_of_the_week using this query

```python
SELECT 
  day_of_the_week, 
  ROUND(AVG(TotalSteps),2) AS avg_step,
  ROUND(AVG(TotalDistance),2) AS avg_dist, 
  ROUND(AVG(Calories),2) AS calories
 FROM [bbt].[dailyactivity_merged]
GROUP BY day_of_the_week
ORDER BY 2 DESC
```

We find that Thursday registered the highest total steps on average for all the activities recorded.

Then I started to take a look at the daily activity, hourly calories, and hourly intensities table. Since the results in the hourly calories table are the same as that in the dailyactivity table as shown by the results of the query below, I decided to use the daily activity table to compare calories burned to workout intensity.

```python
SELECT dailyactivity.id AS dailyactivity_id,
calories.id AS calories_id,
COUNT(dailyactivity.calories) AS dailyactivity_calories,
COUNT(calories.calories) AS total_hourly_calories
FROM [bbt].[dailyactivity_merged] dailyactivity
LEFT JOIN [bbt].[hourlycalories_merged] calories
ON dailyactivity.id = calories.id
GROUP BY dailyactivity.id, calories.id
```

By doing this, I found that the results were the same. showing that the intensity of the exercise equaled the calories burned.

```python
SELECT dailyactivity.id AS dailyactivity_id,
intensity.id AS intensity_id,
COUNT(dailyactivity.calories) AS total_dailyactivity_calories,
COUNT(TotalIntensity) AS totalintensity
FROM [bbt].[dailyactivity_merged] dailyactivity
RIGHT JOIN [bbt].[hourlyintensities_merged] intensity
ON dailyactivity.id = intensity.id
GROUP BY  dailyactivity.id, intensity.id
```

According to an article published by the sleep foundation and authored by
Dr. Elizabeth Rausch-Phung et al, “Most healthy people fall asleep within 15 to 20 minutes of lying down.” I wanted to take a look at the relationship between sleep and physical activity using the total steps column in the daily activity table so I queried the tables using the JOIN statement

```python
SELECT sd.id,
AVG(dailyactivity.totalsteps) AS avg_totalsteps,
dailyactivity.day_of_the_week AS day_of_the_week,
ROUND(AVG(sd.totaltimeinbed),2) AS avg_time_in_bed,
ROUND(AVG(sd.totalminutesasleep),2) avg_time_asleep,
ROUND(AVG(sd.totaltimeinbed - sd.totalminutesasleep),2) avg_time_awake
FROM [bbt].[sleepday_merged] sd
INNER JOIN [bbt].[dailyactivity_merged] dailyactivity
ON dailyactivity.id = sd.id
GROUP BY sd.id, dailyactivity.day_of_the_week
ORDER BY 2 DESC
```

I found that the average time awake, 38 minutes, overshot what is seen as healthy indicating that for these users physical activity did not impact how long it took them to fall asleep. Then I switched it around and ordered by average time awake to see what the highest and lowest average time awake would have about total steps.

```python
SELECT sd.id,
AVG(dailyactivity.totalsteps) AS avg_totalsteps,
dailyactivity.day_of_the_week AS day_of_the_week,
ROUND(AVG(sd.totaltimeinbed),2) AS avg_time_in_bed,
ROUND(AVG(sd.totalminutesasleep),2) avg_time_asleep,
ROUND(AVG(sd.totaltimeinbed - sd.totalminutesasleep),2) avg_time_awake
FROM [bbt].[sleepday_merged] sd
INNER JOIN [bbt].[dailyactivity_merged] dailyactivity
ON dailyactivity.id = sd.id
GROUP BY sd.id, dailyactivity.day_of_the_week
ORDER BY 6 DESC
```

On doing this, I found that the highest average time awake was 371 minutes and the lowest was 1 minute but the total steps for the former was an average of 1551 steps indicating a sedentary lifestyle and the latter was 11997.5 indicating an active lifestyle according to the categorization of Catrine Tudor-Locke et al in their How many steps/day are enough?

This does not correlate with the study done by Alycia N. Sullivan Bisson et al, (2019) in the article — Walk to a Better Night of Sleep: Testing the Relationship Between Physical Activity and Sleep where they concluded that “The current study suggests that PA and sleep were related in multiple ways. The more active one is over a month, the better they report sleeping.” This is regardless of not having data to support sleep quality for this analysis.

Next we see, Total number of times it wasn't worn: This would give you a general sense of how consistent users have been with wearing it. 
* It seems like, users with 1440 minutes of sedantary time and 0 steps in the day did not wear the Fitbit for that day.
